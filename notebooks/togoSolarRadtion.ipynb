#Time series Analysis

```python
# Plotting the distribution of Global Horizontal Irradiance (GHI)
import seaborn as sn
plt.figure(figsize=(10, 6))
sn.histplot(df['GHI'], bins=30, kde=True)
plt.title('Distribution of Global Horizontal Irradiance (GHI)')
plt.xlabel('GHI (W/m)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the relationship between GHI and DNI
plt.figure(figsize=(10, 6))
sn.scatterplot(x='GHI', y='DNI', data=df)
plt.title('GHI vs. DNI')
plt.xlabel('GHI (W/m)')
plt.ylabel('DNI (W/m)')
plt.show()
```

```python

# Time series plot for GHI
plt.figure(figsize=(12, 6))
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
plt.plot(df['Timestamp'], df['GHI'], label='GHI')
plt.xlabel('Time')
plt.ylabel('GHI (W/m)')
plt.title('Time Series of Global Horizontal Irradiance (GHI)')
plt.legend()
plt.show()
```

```python
#plot histogram for GHI, DNI, DHI, WS, and temperatures to visualize the frequency distribution of these variables.
plt.hist(df.GHI, bins =20, rwidth=0.8 )
plt.xlabel("GHI")
plt.ylabel('count')
plt.show()

plt.hist(df.DNI, bins =20, rwidth=0.8 )
plt.xlabel("DNI")
plt.ylabel('count')
plt.show()

plt.hist(df.DHI, bins =20, rwidth=0.8 )
plt.xlabel("DHI")
plt.ylabel('count')
plt.show()

plt.hist(df.WS, bins =20, rwidth=0.8 )
plt.xlabel("WS")
plt.ylabel('count')
plt.show()
```

```python
#check the data is normal distrbution or not
from scipy.stats import norm
plt.hist(df.WS, bins =20, rwidth=0.8, density=True )
plt.xlabel("WS")
plt.ylabel('count')

rng=np.arange(df.WS.min(), df.WS.max(), 0.1)
plt.plot(rng, norm.pdf(rng, df.WS.min(), df.WS.std()))
plt.show()
```
